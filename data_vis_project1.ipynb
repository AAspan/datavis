11. List the average property price for each Neighbourhood. In other words, create a dataframe with two columns: "neighbourhood" and "average_price". (10 points)

```python
import pandas as pd
df = pd.read_excel('airbnb.xlsx', header=0, sheet_name='AirBnB_NYC',usecols=['Neighbourhood ', 'Price'])
df['Average_price'] = df.mean(axis=1)
df = df.set_index(["Neighbourhood ", "Price"])
grouped = df.groupby(level=df.index.names.difference(["Price"]))
grouped.mean()
```

### Summer Olympic Medalists dataset

Load Summer_Olympic_medalists_1896-2008.xslx and answer the following questions. (30 points)

12. From 1980 to 2008, which country won the most medals in Wrestling? (10 points)

```python
import pandas as pd
df = pd.read_excel("Summer_Olympic_medallists_1896-2008.xlsx ")
df1 = df[['Sport','NOC']][(df['Edition'].isin(range(1980,2009))) & (df['Sport'] == 'Wrestling')]
counts = df1.groupby(['NOC'])['Sport'].count().to_dict()
country = max(counts.items(), key = lambda x: x[1])
print("country won the most medals in Wrestling:",country[0])

```

13. List the number of medals won by USA athletes from 1996 to 2008. In other words, create a dataframe with the following columns: "country", "year", "gold", "silver", "bronze". (10 points)

```python
import pandas as pd
df = pd.read_excel("Summer_Olympic_medallists_1896-2008.xlsx ")
df2 = df[["NOC",'Edition','Medal']][(df['Edition'].isin(range(1996,2009))) & (df['NOC'] == 'USA')]
c = df2.groupby(['Edition','Medal'])['Medal'].count().unstack()
c['Country'] = 'USA'
print(c)
```

14. Are there athletes who won medals in three or more different Olympics? If so, print the names, their countries, and the years they won. In other words, create a dataframe with the columns: "athlete", "NOC", "year1", "year2", "year3", "year4", etc. (10 points) 

```python
import pandas as pd
df = pd.read_excel("Summer_Olympic_medallists_1896-2008.xlsx ")
d = df.groupby(['Edition','Athlete','NOC'])['Athlete'].count()
df_counts = d.unstack(0,fill_value=0)
a = list(df_counts)
for col in a:
    df_counts.loc[df_counts[col] > 0, col] = 1
    df_counts.loc[df_counts[col] < 1, col] = 0
df_counts['total'] = df_counts[a].sum(axis = 1)
three_df = df_counts[df_counts['total'] >= 3]
three_df
```
