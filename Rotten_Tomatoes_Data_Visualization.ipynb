# ‚óè	 Films Suggestions:
This plot will utilize user selected sub-categories to find the film with the highest and lowest rated films. For example, a user can select sub-categories,i.e, genre, year, and/ or director, to find the highest or lowest rated movie that meets those specifications. 

```python
import pandas as pd
import numpy as np
import plotly.express as px

df= pd.read_csv("rotten_tomatoes_movies.csv")

highestrate = df.sort_values(by='tomatometer_rating', ascending =False)

newdf=highestrate[['movie_title','movie_info','content_rating','genres','directors','production_company','tomatometer_rating','tomatometer_count','tomatometer_fresh_critics_count']]
                
topTen=newdf.iloc[0:20,:]


fig = px.histogram(topTen, x="tomatometer_count", y="tomatometer_fresh_critics_count", color="genres", marginal="box",
                   hover_data=["movie_title","movie_info"])
fig.show()

fig2 = px.histogram(topTen, x="tomatometer_count", y="tomatometer_fresh_critics_count", color="directors", marginal="box",
                   hover_data=["movie_title","movie_info"])
fig2.show()

fig3 = px.histogram(topTen, x="tomatometer_count", y="tomatometer_fresh_critics_count", color="content_rating", marginal="box",
                   hover_data=["movie_title","movie_info"])
fig3.update_layout(title_text='idk')
fig3.show()

topHigh =newdf.iloc[0:4,:]
fig3 = px.histogram(topHigh, x="tomatometer_count", y="tomatometer_fresh_critics_count", color="production_company", marginal="box",
                   hover_data=["movie_title","movie_info"])
fig3.show()



```


# ‚óè Find My Audience 
This plot is for movie producers who want to determine if their movie will bring in a large audience and to determine what kinds of movies bring in the largest audiences.


```python
import pandas as pd
import numpy as np
import plotly.express as px

df= pd.read_csv("rotten_tomatoes_movies.csv")
df.dropna()

highestrate = df.sort_values(by='tomatometer_rating', ascending =False)

newdf=highestrate[['movie_title','movie_info','content_rating','runtime','genres','directors','production_company',
                   'original_release_date','audience_rating','audience_count','audience_status']]
upRight=newdf.loc[newdf["audience_status"] == "Upright"]    

topTen=upRight.iloc[0:10,:]

fig = px.scatter(topTen, x="audience_rating", y="audience_count",    size="audience_rating", color="genres",
                  size_max=20)
fig.show()

topHigh =newdf.iloc[0:4,:]
fig = px.scatter(topHigh, x="audience_rating", y="audience_count",    size="audience_rating", color="production_company",
                  size_max=5)

fig.show()
fig = px.scatter(topTen, x="audience_rating", y="audience_count",    size="audience_rating", color="directors",
                  size_max=20)

fig.show()

fig = px.scatter(topTen, x="audience_rating", y="audience_count",    size="audience_rating", color="content_rating",
                  size_max=20)

fig.show()
fig = px.scatter(topTen, x="audience_rating", y="audience_count",    size="audience_rating", color="content_rating",
                  size_max=20)

fig.show()
"""group by [audience_status=uplift][] """""
"""remove rows with spilled or no audience status"""
"""top 10"""
"""FacetGrid"""
"""tool tip = "movie_info"""
"""x axis="[audience_status||audience_count]"""
"""Y axis = (interactive), year, genre, actor,contnent rating3 etc."""
```
